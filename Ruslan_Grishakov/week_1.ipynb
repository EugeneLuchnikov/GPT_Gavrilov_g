## Транскрибация

### Видео с tg-канала

```python
# Путь к исходному материалу
source_path = my_drv_path + 'Герман Гаврилов' + '/Видео с tg-канала/'
```

```python
# Путь для файлов транскрибации
transcription_path = my_drv_path + 'Герман Гаврилов' + '/Транскрибация видео с tg-канала/'
```

```python
# Создаём папку для файлов транскрибации
os.makedirs(transcription_path, exist_ok=True)
```

```python
# Осуществляем транскрибацию
import time
import whisper
from tqdm import trange
start_time = time.time()
options = dict(language='ru', beam_size=5, best_of=5)
transcribe_options = dict(task='transcribe', **options)
source_files = os.listdir(source_path)[:-1]
print('Загрузка транскрибатора:', end=' ')
whisper_model = whisper.load_model('large')
for i in trange(len(source_files), desc='Транскрибация'):
  with open(transcription_path + source_files[i].rstrip('.mp4') + '_whisper_large.txt','w') as f:
    f.write(whisper_model.transcribe(source_path + source_files[i], **transcribe_options)['text'])
processing_time = int(time.time() - start_time)
print(f'\nВремя транскрибации: {int(processing_time//60)} мин {int(processing_time%60)} сек.')
```

```python
# Смотрим результат
for filename in os.listdir(transcription_path):
  if filename[0] != '.':
    with open(transcription_path + filename, 'r') as f:
      print(filename)
      print(f.read())
      print()
```

```python
# Большой файл
start_time = time.time()
with open(transcription_path + os.listdir(source_path)[-1].rstrip('.mp4') + '_whisper_large.txt','w') as f:
  f.write(whisper_model.transcribe(source_path + os.listdir(source_path)[-1], **transcribe_options)['text'])
processing_time = int(time.time() - start_time)
print(f'\nВремя транскрибации: {int(processing_time//60)} мин {int(processing_time%60)} сек.')
```
