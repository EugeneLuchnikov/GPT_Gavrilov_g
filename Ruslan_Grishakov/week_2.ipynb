```python
import re
import getpass
import requests
from   langchain               import PromptTemplate
from   langchain.chat_models   import ChatOpenAI
from   langchain.text_splitter import RecursiveCharacterTextSplitter
```

### Функции

```python
# Функция для загрузки документа по ссылке из гугл драйв
def load_document_text(url: str) -> str:

  # Extract the document ID from the URL
  match_ = re.search('/document/d/([a-zA-Z0-9-_]+)', url)
  if match_ is None:
    raise ValueError('Invalid Google Docs URL')
  doc_id = match_.group(1)

  # Download the document as plain text
  response = requests.get(f'https://docs.google.com/document/d/{doc_id}/export?format=txt')
  response.raise_for_status()
  text = response.text

  return text
```

```python
# Функция для извлечения стиля
def get_authors_tone_description(how_to_describe_tone, exam):
  template = """
        Вы — бот с искусственным интеллектом, который очень хорошо умеет писать в одном тоне в качестве примеров.
        Высказывайте свое мнение и обладайте активной позицией.
        Займите твердую позицию в ответе.

        % КАК ОПИСАТЬ ТОН
        {how_to_describe_tone}

        % НАЧАЛО ПРИМЕРОВ
        {post_examples}
        % КОНЕЦ ПРИМЕРОВ

        Перечислите качества стиля приведенных выше примеров.
        """

  prompt = PromptTemplate(input_variables=['how_to_describe_tone', 'post_examples'],
                          template=template,)

  final_prompt = prompt.format(how_to_describe_tone=how_to_describe_tone, post_examples=exam)

  authors_tone_description = llm.predict(final_prompt)

  return authors_tone_description
```

```python
# Функция для определения авторов, которые, по мнению модели, озвучивают примеры постов
def get_similar_public_figures(post_examples):
  template = """
        Вы - ИИ-бот, который умеет определять авторов, общественных деятелей или писателей, чей стиль совпадает с текстом.
        Ваша задача - определить, какие авторы, общественные деятели или писатели наиболее похожи по стилю на приведенный ниже текст

        % НАЧАЛО ПРИМЕРОВ
        {post_examples}
        % КОНЕЦ ПРИМЕРОВ

        Какие авторы (при необходимости перечислите не более 4) наиболее похожи на приведенные выше примеры? Укажите только имена, разделенные запятыми
        """

  prompt = PromptTemplate(input_variables=['post_examples'],
                          template=template,)

  # Using the short list of examples so save on tokens and (hopefully) the top tweets
  final_prompt = prompt.format(post_examples=exam)

  authors = llm.predict(final_prompt)

  return authors
```

```python
# gpt-4
```

```python
# Соберём всё вместе в одной подсказке, обратите внимание на двухшаговый процесс в разделе ВАШЕ ЗАДАНИЕ
template = """
% ИНСТРУКЦИИ
 - Вы - ИИ-бот, который очень хорошо имитирует стиль письма автора.
 - Ваша цель - написать контент в стиле, который описан ниже.
 - Не выходите за рамки приведенных ниже инструкций по стилю
 - При необходимости вставляй ємодзи

% Подражайте стилю этих авторов:
{authors}

% Описание авторского тона:
{tone}

% Образцы авторского стиля письма
{example_text}
% Конец образцов авторского стиля письма

% ВАШЕ ЗАДАНИЕ
1-е - Выпишите темы, о которых может говорить этот автор.
2-е - Напишите пост, как если бы вы были автором, описанным выше
"""

method_4_prompt_template = PromptTemplate(input_variables=["authors", "tone", "example_text"],
                                          template=template,)

# Using the short list of examples so save on tokens and (hopefully) the top tweets
final_prompt = method_4_prompt_template.format(authors=authors,
                                               tone=authors_tone_description,
                                               example_text=exam)
```
